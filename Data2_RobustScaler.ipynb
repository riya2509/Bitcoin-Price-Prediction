### Importing libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Reading the data

```python
df=pd.read_csv('bitcoin_price_Training(data2).csv')
df.head()
```

### Dropping the date column

```python
df_new=df.iloc[:,1:]
```

```python
df_new
```

### Removing commas from 'Volume' column

```python
df_new['Volume'] = df_new['Volume'].str.replace(',', '')
df_new
```

### Removing commas from 'Market cap' column

```python
df_new['Market Cap']=df_new['Market Cap'].str.replace(',','')
```

```python
df_new
```

### Converting '-' values to NaN

```python
df_new['Volume']= df_new['Volume'].replace('-', np.nan)
df_new
```

```python
df_new.info()
```

### Checking for NULL values

```python
df_new.isnull()
```

```python
df_new.isnull().values.any()
```

### Filling missing values

```python
updated_df = df_new
updated_df['Volume']=updated_df['Volume'].fillna(updated_df['Volume'].median())
updated_df.info()
updated_df
```

## Applying Autoencoder

```python
import math
import pandas as pd
import tensorflow as tf
import numpy as np
import kerastuner.tuners as kt
import matplotlib.pyplot as plt
from tensorflow.keras import Model
from tensorflow.keras import Sequential
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split
from tensorflow.keras.losses import MeanSquaredLogarithmicError
from sklearn.preprocessing import PowerTransformer,MinMaxScaler,StandardScaler, RobustScaler
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(updated_df)
data_scaled=scaler.transform(updated_df)
```

```python
type(data_scaled)
```

```python
class Encoder(Model):


  def __init__(self):
    super().__init__()
    self.encoder = Sequential([
      Dense(6, activation='relu'),
      Dense(3, activation='relu'),
      Dense(1, activation='relu'),
      Dropout(0.2),
      Dense(1, activation='relu')
    ])
  def call(self, inputs):
    encoded = self.encoder(inputs)
    return encoded
   
```

```python
model = Encoder()
# configurations of model
model.compile(loss='msle', metrics=['mse'], optimizer='adam')
```

```python
latent_df=model.call(data_scaled)
```

```python
type(latent_df)
```

```python
#latentdim_df=pd.DataFrame(data_scaled.numpy())
latentdim_df=pd.DataFrame(latent_df.numpy())
```
