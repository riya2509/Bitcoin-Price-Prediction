```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('btc_data.csv')
df.head()
```

```python
df_new=df.iloc[:,1:]
```

```python
df_new
```

```python
df1=df_new.reset_index(drop=True)
X=df1.drop('priceUSD',1)
X
```

```python
y=df1[['priceUSD']]
y
```

```python
#Splitting the dataset into the Training set and Test set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train.shape
```

### Building an ANN

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LeakyReLU,PReLU,ELU
from tensorflow.keras.layers import Dropout
```

```python
# Initialising the ANN
classifier = Sequential()
```

```python
# Adding the input layer
classifier.add(Dense(units=735,activation='relu'))
```

```python
# Adding the input layer and the first hidden layer
#2/3*735=units(No of nuerons)
classifier.add(Dense(units=490,activation='relu'))
```

```python
#Adding the second hidden layer
classifier.add(Dense(units=400,activation='relu'))
```

```python
#Adding the output layer
classifier.add(Dense(1,activation='linear'))
```

```python
classifier.compile(optimizer='adam',loss='mean_absolute_error',metrics=['mse'])
```

```python
model_history=classifier.fit(X_train,y_train,validation_split=0.33,batch_size=10,epochs=50)
```

```python
# evaluate the model
train_mse = classifier.evaluate(X_train, y_train, verbose=0)
test_mse = classifier.evaluate(X_test, y_test, verbose=0)
```

```python
# plot loss during training
import matplotlib.pyplot as plt
plt.title('Mean Absolute Error Loss')
plt.plot(model_history.history['loss'], label='train')
plt.plot(model_history.history['val_loss'], label='test')
plt.legend()
plt.show()
```
