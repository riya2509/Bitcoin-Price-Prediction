```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('btc_data.csv')
df.head()
```

```python
df_new=df.iloc[:,1:]
```

```python
df_new
```

```python
df1=df_new.reset_index(drop=True)
X=df1.drop('priceUSD',1)
X
```

```python
y=df1[['priceUSD']]
y
```

```python
#Splitting the dataset into the Training set and Test set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

### Building an ANN

```python
import keras
```

```python
# Building the neural network layer by layer
from keras.models import Sequential
```

```python
#To randomly initialize the weights to small numbers close to 0(But not 0)
from keras.layers import Dense
```

```python
classifier = Sequential()
```

```python
classifier.add(Dense(output_dim = 6, init = 'uniform', units=64,activation = 'relu', input_dim = 11))
```

```python
#Second hidden layer
classifier.add(Dense(output_dim = 6, units=64,init = 'uniform', activation = 'relu'))
```

```python
#Adding the output layer
classifier.add(Dense(output_dim = 1,units=64, init = 'uniform', activation = 'sigmoid'))
```

```python
#Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Fitting the ANN to the training set
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

```python
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
#if y_pred is larger than 0.5 it returns true(1) else false(2)
print(y_pred)
```

```python
# Confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score)
```
