### Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from pycaret.regression import *
```

### Reading the data and storing it into a variable

```python
asset=pd.read_csv('btc_data.csv')
```

```python
#Showing the data
asset
```

```python
# A variable for predicting one day out into the future
future_days=1
#Creating a new column (target or dependent variable) shifts n=1 units up
asset['Future_Price']= asset[['priceUSD']].shift(-future_days)
# Choosing the columns to keep
asset = asset[['priceUSD','Future_Price']]
asset
```

```python
#Making copy of dataset
df=asset.copy()
#Creating independent dataset
X=np.array(df[df.columns])
#removing the last 'n' rows from the datset where 'n' is future_days=1
X=X[:len(asset)-future_days]
#Creating the dependent dataset y and coverting dataframe into numpy array
y=np.array(df['Future_Price'])
#Getting all the y values except the last 'n' rows
y=y[:-future_days]
```

### Splitting into 85% training and 15% testing data set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.15,random_state=0,shuffle=False)
```

### Applying kernel PCA

```python
from sklearn.decomposition import KernelPCA
kpca=KernelPCA(n_components=0.95,kernel='rbf')
X_train=kpca.fit_transform(X_train)
X_test=kpca.transform(X_test)
```

```python
#Getting the train data and transforming into dataframe
train_data=pd.DataFrame(X_train,columns=df.columns)
train_data.head(7)
```

```python
#transforming test data into dataframe
test_data=pd.DataFrame(X_test,columns=df.columns)
test_data.head(7)
```
